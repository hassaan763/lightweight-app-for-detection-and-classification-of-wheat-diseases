```python
import os
import torch
import torch.nn as nn
from torchvision import datasets, transforms, models
from torchvision.models import MobileNet_V3_Small_Weights
from torch.utils.data import DataLoader, WeightedRandomSampler
from tqdm import tqdm
import matplotlib.pyplot as plt
from torchsummary import summary
from sklearn.metrics import confusion_matrix, classification_report
import numpy as np
import warnings

# Suppress specific warnings
warnings.filterwarnings("ignore", category=UserWarning)


class EarlyStopping:
    """Early stopping utility to prevent overfitting and escape local maxima."""
    
    def __init__(self, patience=7, min_delta=0.001, restore_best_weights=True):
        """
        Args:
            patience (int): How many epochs to wait after last time validation accuracy improved
            min_delta (float): Minimum change in the monitored quantity to qualify as an improvement
            restore_best_weights (bool): Whether to restore model weights from the epoch with the best value
        """
        self.patience = patience
        self.min_delta = min_delta
        self.restore_best_weights = restore_best_weights
        self.best_acc = 0
        self.counter = 0
        self.best_weights = None
        self.early_stop = False
        
    def __call__(self, val_acc, model):
        """
        Check if training should stop early.
        
        Args:
            val_acc (float): Current validation accuracy
            model: PyTorch model
            
        Returns:
            bool: True if training should stop
        """
        if val_acc > self.best_acc + self.min_delta:
            self.best_acc = val_acc
            self.counter = 0
            if self.restore_best_weights:
                self.best_weights = model.state_dict().copy()
        else:
            self.counter += 1
            
        if self.counter >= self.patience:
            self.early_stop = True
            if self.restore_best_weights and self.best_weights is not None:
                model.load_state_dict(self.best_weights)
                print(f"üîÑ Restored model weights from best epoch (accuracy: {self.best_acc:.2f}%)")
            
        return self.early_stop


class WheatDiseaseTrainer:
    def __init__(self, base_dir="split_dataset", num_classes=7):
        """
        Initialize the wheat disease trainer.
        
        Args:
            base_dir (str): Base directory containing train/val/test folders
            num_classes (int): Number of disease classes
        """
        self.base_dir = base_dir
        self.num_classes = num_classes
        self.device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
        print(f"Using device: {self.device}")
        
        # Print GPU info if available
        if torch.cuda.is_available():
            print(f"CUDA Device: {torch.cuda.get_device_name()}")
            print(f"CUDA Memory: {torch.cuda.get_device_properties(0).total_memory / 1e9:.1f} GB")
            torch.cuda.empty_cache()
        
        # ImageNet normalization values
        self.mean = [0.485, 0.456, 0.406]
        self.std = [0.229, 0.224, 0.225]
        
    def _validate_directories(self):
        """Validate that required directories exist."""
        required_dirs = ['train', 'val', 'test']
        for dir_name in required_dirs:
            dir_path = os.path.join(self.base_dir, dir_name)
            if not os.path.exists(dir_path):
                raise FileNotFoundError(f"Required directory not found: {dir_path}")
                
    def _get_transforms(self):
        """Create data transforms for training and validation/testing."""
        train_transforms = transforms.Compose([
            transforms.Resize((224, 224)),
            transforms.RandomHorizontalFlip(p=0.5),
            transforms.RandomRotation(15),
            transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2),
            transforms.ToTensor(),
            transforms.Normalize(self.mean, self.std)
        ])

        val_test_transforms = transforms.Compose([
            transforms.Resize((224, 224)),
            transforms.ToTensor(),
            transforms.Normalize(self.mean, self.std)
        ])
        
        return train_transforms, val_test_transforms
    
    def _get_class_weights(self, dataset):
        """
        Calculate class weights for handling imbalanced datasets.
        
        Args:
            dataset: ImageFolder dataset
            
        Returns:
            tuple: (sample_weights, class_weights)
        """
        # Count samples per class
        class_counts = [0] * len(dataset.classes)
        for _, label in dataset.samples:
            class_counts[label] += 1
        
        print(f"Class distribution: {dict(zip(dataset.classes, class_counts))}")
        
        # Compute weights: inverse frequency
        total_samples = sum(class_counts)
        class_weights = []
        for count in class_counts:
            weight = total_samples / (len(dataset.classes) * count)
            class_weights.append(weight)
        
        # Create sample weights for WeightedRandomSampler
        sample_weights = []
        for _, label in dataset.samples:
            sample_weights.append(class_weights[label])
        
        return sample_weights, class_weights
    
    def _create_model(self):
        """Create and configure the MobileNetV3 model."""
        # Load pretrained model with new weights parameter (fixes deprecation warning)
        model = models.mobilenet_v3_small(weights=MobileNet_V3_Small_Weights.IMAGENET1K_V1)
        
        # Freeze early layers for transfer learning
        for param in model.features[:-3].parameters():
            param.requires_grad = False
            
        # Replace classifier
        model.classifier[3] = nn.Linear(model.classifier[3].in_features, self.num_classes)
        
        return model.to(self.device)
    
    def _save_class_labels(self, class_names, label_file="class_labels.txt"):
        """Save class labels to file."""
        try:
            with open(label_file, "w") as f:
                for label in class_names:
                    f.write(label + "\n")
            print(f"‚úÖ Saved class labels to '{label_file}': {class_names}")
        except IOError as e:
            print(f"‚ùå Error saving class labels: {e}")
    
    def _plot_training_history(self, train_losses, val_losses, train_accuracies, val_accuracies, num_epochs, stopped_epoch=None):
        """Plot training history."""
        epochs = range(1, len(train_losses) + 1)
        
        plt.figure(figsize=(15, 5))
        
        # Loss plot
        plt.subplot(1, 2, 1)
        plt.plot(epochs, train_losses, 'b-', label='Training Loss', linewidth=2)
        plt.plot(epochs, val_losses, 'r-', label='Validation Loss', linewidth=2)
        if stopped_epoch:
            plt.axvline(x=stopped_epoch, color='orange', linestyle='--', alpha=0.7, label=f'Early Stop (Epoch {stopped_epoch})')
        plt.xlabel('Epoch')
        plt.ylabel('Loss')
        plt.title('Training and Validation Loss')
        plt.legend()
        plt.grid(True, alpha=0.3)
        
        # Accuracy plot
        plt.subplot(1, 2, 2)
        plt.plot(epochs, train_accuracies, 'b-', label='Training Accuracy', linewidth=2)
        plt.plot(epochs, val_accuracies, 'r-', label='Validation Accuracy', linewidth=2)
        if stopped_epoch:
            plt.axvline(x=stopped_epoch, color='orange', linestyle='--', alpha=0.7, label=f'Early Stop (Epoch {stopped_epoch})')
        plt.xlabel('Epoch')
        plt.ylabel('Accuracy (%)')
        plt.title('Training and Validation Accuracy')
        plt.legend()
        plt.grid(True, alpha=0.3)
        
        plt.tight_layout()
        plt.savefig('training_history.png', dpi=300, bbox_inches='tight')
        plt.show()
    
    def evaluate_model(self, model, test_loader, criterion, class_names):
        """
        Evaluate model on test set.
        
        Args:
            model: Trained model
            test_loader: Test data loader
            criterion: Loss function
            class_names: List of class names
            
        Returns:
            tuple: (test_loss, test_accuracy, confusion_matrix)
        """
        model.eval()
        test_loss = 0.0
        correct = 0
        total = 0
        all_preds = []
        all_labels = []

        print("\n" + "="*50)
        print("EVALUATING ON TEST SET")
        print("="*50)

        with torch.no_grad():
            for images, labels in tqdm(test_loader, desc="Testing"):
                try:
                    images, labels = images.to(self.device), labels.to(self.device)
                    outputs = model(images)
                    loss = criterion(outputs, labels)
                    
                    test_loss += loss.item()
                    _, predicted = outputs.max(1)
                    total += labels.size(0)
                    correct += predicted.eq(labels).sum().item()
                    
                    # Store predictions and labels
                    all_preds.extend(predicted.cpu().numpy())
                    all_labels.extend(labels.cpu().numpy())
                    
                except RuntimeError as e:
                    if "out of memory" in str(e):
                        print(f"‚ùå CUDA out of memory during testing. Clearing cache...")
                        torch.cuda.empty_cache()
                        continue
                    else:
                        raise e

        # Calculate metrics
        avg_loss = test_loss / len(test_loader)
        accuracy = (correct / total) * 100

        print(f"\nüìä TEST RESULTS:")
        print(f"   Test Loss: {avg_loss:.4f}")
        print(f"   Test Accuracy: {accuracy:.2f}%")

        # Confusion Matrix
        cm = confusion_matrix(all_labels, all_preds)
        print(f"\nüìà CONFUSION MATRIX:")
        print(cm)

        # Classification Report
        print(f"\nüìã DETAILED CLASSIFICATION REPORT:")
        print(classification_report(all_labels, all_preds, target_names=class_names))

        return avg_loss, accuracy, cm
    
    def train(self, 
              output_model_path="mobilenetv3_wheat_disease.pt",
              label_file="class_labels.txt",
              num_epochs=15,
              batch_size=32,
              learning_rate=0.001,
              use_weighted_sampling=True,
              patience=10,
              min_delta=0.001):
        """
        Train the wheat disease classification model with early stopping.
        
        Args:
            output_model_path (str): Path to save the trained model
            label_file (str): Path to save class labels
            num_epochs (int): Maximum number of training epochs
            batch_size (int): Batch size for training
            learning_rate (float): Learning rate for optimizer
            use_weighted_sampling (bool): Whether to use weighted sampling for class imbalance
            patience (int): Early stopping patience (epochs to wait for improvement)
            min_delta (float): Minimum improvement threshold for early stopping
        """
        try:
            print("üåæ WHEAT DISEASE CLASSIFICATION TRAINING")
            print("="*50)
            
            # Validate directories
            self._validate_directories()
            
            # Get transforms
            train_transforms, val_test_transforms = self._get_transforms()
            
            # Load datasets
            train_dir = os.path.join(self.base_dir, "train")
            val_dir = os.path.join(self.base_dir, "val")
            test_dir = os.path.join(self.base_dir, "test")
            
            print(f"üìÅ Loading datasets from: {self.base_dir}")
            
            try:
                train_dataset = datasets.ImageFolder(train_dir, transform=train_transforms)
                val_dataset = datasets.ImageFolder(val_dir, transform=val_test_transforms)
                test_dataset = datasets.ImageFolder(test_dir, transform=val_test_transforms)
            except Exception as e:
                print(f"‚ùå Error loading datasets: {e}")
                raise
            
            print(f"   Training samples: {len(train_dataset)}")
            print(f"   Validation samples: {len(val_dataset)}")
            print(f"   Test samples: {len(test_dataset)}")
            
            # Adjust num_workers and pin_memory based on system
            num_workers = 2 if self.device.type == 'cuda' else 0
            pin_memory = True if self.device.type == 'cuda' else False
            
            # Handle class imbalance
            if use_weighted_sampling:
                sample_weights, class_weights = self._get_class_weights(train_dataset)
                sampler = WeightedRandomSampler(
                    weights=sample_weights, 
                    num_samples=len(sample_weights), 
                    replacement=True
                )
                train_loader = DataLoader(
                    train_dataset, 
                    batch_size=batch_size, 
                    sampler=sampler, 
                    num_workers=num_workers, 
                    pin_memory=pin_memory
                )
                criterion = nn.CrossEntropyLoss()  # Use regular loss with weighted sampling
                print("‚úÖ Using weighted random sampling for class imbalance")
            else:
                train_loader = DataLoader(
                    train_dataset, 
                    batch_size=batch_size, 
                    shuffle=True, 
                    num_workers=num_workers, 
                    pin_memory=pin_memory
                )
                _, class_weights = self._get_class_weights(train_dataset)
                criterion = nn.CrossEntropyLoss(weight=torch.tensor(class_weights, dtype=torch.float32).to(self.device))
                print("‚úÖ Using weighted loss function for class imbalance")
            
            val_loader = DataLoader(
                val_dataset, 
                batch_size=batch_size, 
                shuffle=False, 
                num_workers=num_workers, 
                pin_memory=pin_memory
            )
            test_loader = DataLoader(
                test_dataset, 
                batch_size=batch_size, 
                shuffle=False, 
                num_workers=num_workers, 
                pin_memory=pin_memory
            )
            
            # Save class labels
            class_names = train_dataset.classes
            self._save_class_labels(class_names, label_file)
            
            # Create model
            print("\nüß† Creating model...")
            model = self._create_model()
            
            print("\nüìã MODEL ARCHITECTURE SUMMARY:")
            try:
                summary(model, (3, 224, 224))
            except Exception as e:
                print(f"Could not display torchsummary: {e}")
                # Alternative: print basic model info
                total_params = sum(p.numel() for p in model.parameters())
                trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
                print(f"Total parameters: {total_params:,}")
                print(f"Trainable parameters: {trainable_params:,}")
            
            # Initialize early stopping
            early_stopping = EarlyStopping(patience=patience, min_delta=min_delta, restore_best_weights=True)
            print(f"üõ°Ô∏è Early stopping enabled - patience: {patience} epochs, min_delta: {min_delta}")
            
            # Optimizer and scheduler
            optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate, weight_decay=1e-4)
            scheduler = torch.optim.lr_scheduler.StepLR(optimizer, step_size=7, gamma=0.1)
            
            # Training history
            train_losses, val_losses = [], []
            train_accuracies, val_accuracies = [], []
            best_val_acc = 0.0
            stopped_epoch = None
            
            print(f"\nüöÄ Starting training for up to {num_epochs} epochs...")
            print("="*80)
            
            for epoch in range(num_epochs):
                try:
                    # Training phase
                    model.train()
                    running_loss = 0.0
                    correct = 0
                    total = 0
                    
                    progress_bar = tqdm(train_loader, desc=f"Epoch {epoch+1}/{num_epochs}")
                    for batch_idx, (images, labels) in enumerate(progress_bar):
                        try:
                            images, labels = images.to(self.device, non_blocking=True), labels.to(self.device, non_blocking=True)
                            
                            optimizer.zero_grad()
                            outputs = model(images)
                            loss = criterion(outputs, labels)
                            loss.backward()
                            optimizer.step()
                            
                            running_loss += loss.item()
                            _, predicted = outputs.max(1)
                            total += labels.size(0)
                            correct += predicted.eq(labels).sum().item()
                            
                            # Update progress bar
                            current_acc = (correct / total) * 100
                            progress_bar.set_postfix({
                                'Loss': f'{loss.item():.4f}',
                                'Acc': f'{current_acc:.2f}%'
                            })
                            
                        except RuntimeError as e:
                            if "out of memory" in str(e):
                                print(f"\n‚ùå CUDA out of memory at batch {batch_idx}. Try reducing batch_size")
                                print(f"Current batch_size: {batch_size}")
                                torch.cuda.empty_cache()
                                raise e
                            else:
                                raise e
                    
                    train_loss = running_loss / len(train_loader)
                    train_acc = (correct / total) * 100
                    
                    # Validation phase
                    model.eval()
                    val_running_loss = 0.0
                    correct = 0
                    total = 0
                    
                    with torch.no_grad():
                        for images, labels in val_loader:
                            try:
                                images, labels = images.to(self.device, non_blocking=True), labels.to(self.device, non_blocking=True)
                                outputs = model(images)
                                loss = criterion(outputs, labels)
                                val_running_loss += loss.item()
                                
                                _, predicted = outputs.max(1)
                                total += labels.size(0)
                                correct += predicted.eq(labels).sum().item()
                                
                            except RuntimeError as e:
                                if "out of memory" in str(e):
                                    print(f"\n‚ùå CUDA out of memory during validation. Clearing cache...")
                                    torch.cuda.empty_cache()
                                    continue
                                else:
                                    raise e
                    
                    val_loss = val_running_loss / len(val_loader)
                    val_acc = (correct / total) * 100
                    
                    # Update scheduler
                    scheduler.step()
                    
                    # Store metrics
                    train_losses.append(train_loss)
                    val_losses.append(val_loss)
                    train_accuracies.append(train_acc)
                    val_accuracies.append(val_acc)
                    
                    # Print epoch results
                    print(f"Epoch [{epoch+1}/{num_epochs}] "
                          f"Train Loss: {train_loss:.4f} | Val Loss: {val_loss:.4f} "
                          f"| Train Acc: {train_acc:.2f}% | Val Acc: {val_acc:.2f}% "
                          f"| LR: {scheduler.get_last_lr()[0]:.2e}")
                    
                    # Save best model (separate from early stopping for backup)
                    if val_acc > best_val_acc:
                        best_val_acc = val_acc
                        try:
                            torch.save({
                                'epoch': epoch,
                                'model_state_dict': model.state_dict(),
                                'optimizer_state_dict': optimizer.state_dict(),
                                'val_acc': val_acc,
                                'class_names': class_names
                            }, output_model_path)
                            print(f"   ‚úÖ New best model saved! Validation Accuracy: {val_acc:.2f}%")
                        except Exception as e:
                            print(f"‚ùå Error saving model: {e}")
                    
                    # Check early stopping
                    if early_stopping(val_acc, model):
                        stopped_epoch = epoch + 1
                        print(f"üõë Early stopping triggered after {stopped_epoch} epochs")
                        print(f"   Best validation accuracy: {early_stopping.best_acc:.2f}%")
                        print(f"   No improvement for {patience} consecutive epochs")
                        break
                    
                    # Print early stopping status
                    if early_stopping.counter > 0:
                        print(f"   ‚è≥ Early stopping counter: {early_stopping.counter}/{patience}")
                    
                    # Clear cache after each epoch to prevent memory buildup
                    if self.device.type == 'cuda':
                        torch.cuda.empty_cache()
                        
                except Exception as e:
                    print(f"‚ùå Error in epoch {epoch+1}: {e}")
                    if "out of memory" in str(e):
                        print("üí° Suggestions:")
                        print("   1. Reduce batch_size (try 16 or 8)")
                        print("   2. Reduce image resolution")
                        print("   3. Use CPU instead of GPU")
                    raise e
            
            # Final model selection
            final_acc = early_stopping.best_acc if early_stopping.early_stop else best_val_acc
            
            print(f"\nüéâ Training completed!")
            if stopped_epoch:
                print(f"   Training stopped early at epoch {stopped_epoch}")
                print(f"   Final validation accuracy: {final_acc:.2f}%")
            else:
                print(f"   Completed all {num_epochs} epochs")
                print(f"   Best validation accuracy: {best_val_acc:.2f}%")
            print(f"   Model saved to: {output_model_path}")
            
            # Plot training history
            self._plot_training_history(train_losses, val_losses, train_accuracies, val_accuracies, num_epochs, stopped_epoch)
            
            # Load best model for final evaluation
            try:
                checkpoint = torch.load(output_model_path, map_location=self.device)
                model.load_state_dict(checkpoint['model_state_dict'])
                
                # Final test evaluation
                test_loss, test_acc, cm = self.evaluate_model(model, test_loader, criterion, class_names)
                
                print(f"\nüèÜ FINAL RESULTS:")
                print(f"   Best Validation Accuracy: {final_acc:.2f}%")
                print(f"   Final Test Accuracy: {test_acc:.2f}%")
                if stopped_epoch:
                    print(f"   Training stopped early at epoch: {stopped_epoch}")
                
                return model, test_acc, cm
                
            except Exception as e:
                print(f"‚ùå Error loading best model for evaluation: {e}")
                return model, 0.0, None
                
        except Exception as e:
            print(f"‚ùå Critical error during training: {e}")
            raise


def main():
    """Main function to run the training."""
    # Configuration with early stopping parameters
    config = {
        'base_dir': '/home/hassaan-ahmad/aug_data_train_test',  # Change this to your dataset path
        'output_model_path': '/home/hassaan-ahmad/aug_data_train_test/aug_mobilenetv3_wheat_disease_best.pt',
        'label_file': 'wheat_class_labels.txt',
        'num_epochs': 50,  # Maximum epochs - training may stop early
        'batch_size': 128,
        'learning_rate': 0.001,
        'num_classes': 7,
        'use_weighted_sampling': True,
        'patience': 10,  # Wait 10 epochs for improvement before stopping
        'min_delta': 0.001  # Minimum improvement threshold (0.1%)
    }
    
    try:
        print("üåæ STARTING WHEAT DISEASE CLASSIFICATION PIPELINE")
        print("="*60)
        
        # Initialize trainer
        trainer = WheatDiseaseTrainer(
            base_dir=config['base_dir'], 
            num_classes=config['num_classes']
        )
        
        # Train model with early stopping
        model, test_acc, confusion_matrix = trainer.train(
            output_model_path=config['output_model_path'],
            label_file=config['label_file'],
            num_epochs=config['num_epochs'],
            batch_size=config['batch_size'],
            learning_rate=config['learning_rate'],
            use_weighted_sampling=config['use_weighted_sampling'],
            patience=config['patience'],
            min_delta=config['min_delta']
        )
        
        print(f"\n‚úÖ Training pipeline completed successfully!")
        print(f"   Final test accuracy: {test_acc:.2f}%")
        print(f"   Model saved to: {config['output_model_path']}")
        print(f"   Class labels saved to: {config['label_file']}")
        
    except KeyboardInterrupt:
        print("\n‚ö†Ô∏è Training interrupted by user (Ctrl+C)")
        print("   You can resume training by running the script again")
    except RuntimeError as e:
        if "out of memory" in str(e):
            print(f"\n‚ùå CUDA out of memory error: {e}")
            print("\nüí° Try these solutions:")
            print("   1. Reduce batch_size to 16 or 8")
            print("   2. Reduce image size from 224 to 192 or 128")
            print("   3. Use CPU by setting device='cpu'")
            print("   4. Close other GPU-using applications")
        else:
            print(f"‚ùå Runtime error: {e}")
            raise
    except Exception as e:
        print(f"‚ùå Unexpected error during training: {str(e)}")
        raise


if __name__ == "__main__":
    main()
```
